```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from config import *
from langchain_openai import AzureOpenAIEmbeddings

# Initialize the embeddings client
embedding_client = AzureOpenAIEmbeddings(
    azure_deployment=EMBEDDING_DEPLOYMENT,
    openai_api_version="2023-05-15",
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    api_key=AZURE_OPENAI_KEY,
    chunk_size=1
)

# Test with obviously similar queries
test_queries = [
    "500 error while calling login API",
    "500 internal server error", 
    "login API failed with 500",
    "database connection timeout",  # Different topic
    "user authentication successful"  # Very different
]

print("ðŸ” Testing embedding similarities...")
print("Generating embeddings for test queries...")

vectors = []
for query in test_queries:
    vector = np.array(embedding_client.embed_documents([query])).astype('float32')
    vectors.append(vector[0])
    print(f"âœ… '{query}' -> vector shape: {vector.shape}")

print(f"\nðŸ“Š All vectors shape: {len(vectors)} vectors, each with {len(vectors[0])} dimensions")

# Calculate similarity matrix
similarity_matrix = cosine_similarity(vectors)

print("\nðŸŽ¯ COSINE SIMILARITIES BETWEEN TEST QUERIES:")
print("=" * 80)

for i, query1 in enumerate(test_queries):
    for j, query2 in enumerate(test_queries):
        if i < j:
            similarity = similarity_matrix[i][j]
            print(f"'{query1}'")
            print(f"  vs '{query2}'")
            print(f"  Similarity: {similarity:.4f}")
            print("-" * 60)

print("\nðŸ“ˆ SIMILARITY MATRIX:")
print("Rows/Columns: ", [f"Q{i+1}" for i in range(len(test_queries))])
print(similarity_matrix.round(4))
```

```python
# ===========================================
# ðŸ” Azure OpenAI Embeddings + FAISS Testing
# ===========================================

from langchain_openai import AzureOpenAIEmbeddings
import numpy as np
import faiss
import time
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
from config import *  # Your config file with keys and endpoints

# ------------------------------------------------------------
# 1ï¸âƒ£ Initialize Azure OpenAI Embeddings client
# ------------------------------------------------------------
embedding_client = AzureOpenAIEmbeddings(
    azure_deployment=EMBEDDING_DEPLOYMENT,
    openai_api_version="2023-05-15",
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    api_key=AZURE_OPENAI_KEY,
    chunk_size=1
)

# ------------------------------------------------------------
# 2ï¸âƒ£ Sample texts for testing
# ------------------------------------------------------------
texts = [
    "login API returned 500 error",
    "login endpoint failed with internal server error",
    "user authentication successful",
    "database connection timeout",
    "server responded with 502 bad gateway"
]

# ------------------------------------------------------------
# 3ï¸âƒ£ Generate embeddings
# ------------------------------------------------------------
print("ðŸ”¹ Generating embeddings...")
start = time.time()
embeddings = np.array(embedding_client.embed_documents(texts)).astype('float32')
print(f"âœ… Embeddings generated in {time.time() - start:.2f}s")
print(f"Vector shape: {embeddings.shape}")

# Normalize for cosine similarity
faiss.normalize_L2(embeddings)

# ------------------------------------------------------------
# 4ï¸âƒ£ Create FAISS index and add vectors
# ------------------------------------------------------------
dim = embeddings.shape[1]  # typically 1536 for Azure ada model
index = faiss.IndexFlatIP(dim)  # IP â‰ˆ cosine similarity when normalized
index.add(embeddings)
print(f"ðŸ“¦ Added {index.ntotal} vectors to FAISS index.")

# ------------------------------------------------------------
# 5ï¸âƒ£ Perform semantic search
# ------------------------------------------------------------
query = "login api gives 500 internal error"
query_vector = np.array(embedding_client.embed_documents([query])).astype('float32')
faiss.normalize_L2(query_vector)

start = time.time()
k = 3
scores, indices = index.search(query_vector, k)
elapsed = time.time() - start

print(f"\nðŸ” Query: '{query}'")
print(f"âš¡ Search completed in {elapsed*1000:.2f} ms\n")

for rank, (i, score) in enumerate(zip(indices[0], scores[0]), start=1):
    print(f"{rank}. {texts[i]} (similarity: {score:.4f})")

# ------------------------------------------------------------
# 6ï¸âƒ£ Optional: Visualize similarity between all stored texts
# ------------------------------------------------------------
print("\nðŸ“Š Pairwise cosine similarities among all stored texts:")
sim_matrix = cosine_similarity(embeddings)
print(np.round(sim_matrix, 3))

plt.figure(figsize=(7,6))
sns.heatmap(sim_matrix, annot=True, xticklabels=texts, yticklabels=texts, cmap="YlGnBu")
plt.title("Cosine Similarity Heatmap of FAISS-Stored Embeddings")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# ðŸ” Validate FAISS Store Results (Azure Embeddings + Metadata)
# ============================================================

import faiss
import pickle
import numpy as np
import time
from langchain_openai import AzureOpenAIEmbeddings
from config import *

# ------------------------------------------------------------
# 1ï¸âƒ£ Load FAISS index + metadata
# ------------------------------------------------------------
faiss_index_path = r"C:\log\faiss_http_index\index.faiss"
metadata_path = r"C:\log\faiss_http_index\index.pkl"

print("ðŸ“¦ Loading FAISS index and metadata...")
index = faiss.read_index(faiss_index_path)

with open(metadata_path, "rb") as f:
    texts = pickle.load(f)

print(f"âœ… Index loaded with {index.ntotal} vectors.")
print(f"âœ… Metadata loaded with {len(texts)} text entries.")

# ------------------------------------------------------------
# 2ï¸âƒ£ Initialize Azure OpenAI Embeddings client
# ------------------------------------------------------------
embedding_client = AzureOpenAIEmbeddings(
    azure_deployment=EMBEDDING_DEPLOYMENT,
    openai_api_version="2023-05-15",
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    api_key=AZURE_OPENAI_KEY,
    chunk_size=1
)

# ------------------------------------------------------------
# 3ï¸âƒ£ Query test function
# ------------------------------------------------------------
def search_faiss(query: str, top_k: int = 5):
    """Safely search your FAISS index and display top similar results"""
    print(f"\nðŸ” Query: {query}")
    query_vector = np.array(embedding_client.embed_documents([query])).astype('float32')
    faiss.normalize_L2(query_vector)
    
    start = time.time()
    scores, ids = index.search(query_vector, top_k)
    elapsed = (time.time() - start) * 1000
    
    print(f"âš¡ Search time: {elapsed:.2f} ms\n")
    
    for rank, (i, score) in enumerate(zip(ids[0], scores[0]), start=1):
        # Skip invalid or out-of-range IDs
        if i < 0 or i >= len(texts):
            print(f"{rank}. âš ï¸ Invalid index ID: {i}")
            continue
        
        text_preview = str(texts[i])[:200].replace("\n", " ")
        print(f"{rank}. Index ID: {i}, Similarity: {score:.4f}")
        print(f"   â†’ {text_preview}...\n")

# ------------------------------------------------------------
# 4ï¸âƒ£ Run sample queries to test correctness
# ------------------------------------------------------------
search_faiss("500 error while calling login API")
search_faiss("authentication failed")
search_faiss("database timeout")
```
